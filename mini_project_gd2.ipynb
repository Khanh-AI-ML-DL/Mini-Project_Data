```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import numpy as np
```

```python
df = pd.read_csv('SuperMarket Analysis.csv')
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
doanh_thu_product = (df.groupby('Product line')[ 'cogs' ].sum().sort_values(ascending=False))
doanh_thu_product
```

```python
doanh_thu_branch = (df.groupby('Branch')[ 'cogs' ].sum().sort_values(ascending=False))
doanh_thu_branch
```

```python
time_good = (df.groupby('Time')[ 'Date' ].count().sort_values(ascending=False))
time_good
```

```python
# Bieu do cot
plt.figure(figsize = (10,10))
doanh_thu_product.plot(kind='bar')
plt.xlabel('Product line ')
plt.ylabel('cogs ')
plt.xticks(rotation = 20)
plt.title('Cogs theo  Product line ')
plt.show()
```

```python
# Bieu do cot
doanh_thu_ngay = df.groupby('Date')['cogs'].sum()
plt.figure(figsize = (10,10))
doanh_thu_ngay.plot(kind='line')
plt.xlabel('Cogs')
plt.xlabel('Cogs')
plt.ylabel('Date')
plt.title('Cogs theo  Date')
plt.show()
```

```python
Payment_type_count = df['Payment'].value_counts()
Payment_type_count
plt.pie(Payment_type_count ,labels=Payment_type_count.index , autopct = '%1.1f%%')
plt.title('ti le type  Payment')
plt.show()
```
